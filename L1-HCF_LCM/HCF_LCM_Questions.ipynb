    #### 20. HCF & LCM
Send Feedback
HCF of two expressions is (x + 1) and LCM is (x^3+ x^2 – x – 1). 
If one expression is (x^2 - 1), then what is the second expression?

Options
x + 1
(x - 1)^2
(x + 1)^2 
(x - 1)( x + 1)

Correct Answer
Solution Description
HCF = (x + 1)

LCM = x^3 + x^2 – x – 1 = (x + 1) (x + 1) (x – 1)

First expression = x^2 – 1 = (x + 1) (x - 1).

 Second expression × first expression = HCF x LCM

Second expression= HCF x LCM/first expression 
= ((x+1)(x+1)(x+1)(x-1))/((x-1)(x+1))= (x+1)^2
